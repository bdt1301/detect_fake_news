Giới Thiệu

Trong thời đại mạng xã hội ngày nay, việc đặt niềm tin vào mọi thông tin mà chúng ta nghe được trở nên khó khăn hơn bao giờ hết. Bạn có tin tưởng mọi thông tin từ mạng xã hội hay không? Liệu tất cả tin tức đều là không có thật?

Tin tức giả, hay còn được gọi là tin rác, là những thông tin sai lệch được trình bày dưới dạng tin tức với mục đích gây hiểu lầm hoặc tổn hại danh tiếng cá nhân, tổ chức, thậm chí kiếm lợi nhuận thông qua doanh thu quảng cáo. Thuật ngữ này không chỉ giới hạn ở thông tin mạo danh mà còn bao gồm các thông tin sai lệch do các cơ chế vô ý và vô ý thức, cũng như sự lạm dụng của những người có quyền lực.

Với sự bùng nổ của mạng xã hội, đặc biệt là Facebook News Feed, tin tức giả trở nên phổ biến hơn. Phân cực chính trị, chính trị hậu sự thật, và thuật toán truyền thông xã hội đóng một vai trò quan trọng trong việc lan truyền tin giả. Có cả những nỗ lực của các tác nhân nước ngoài thù địch, đặc biệt là trong các cuộc bầu cử, để tạo và lan truyền tin tức giả.

Tin tức giả không chỉ ảnh hưởng đến sự cạnh tranh với tin tức thật mà còn có thể làm suy giảm lòng tin vào tin tức đáng tin cậy. Một số nghiên cứu đã chỉ ra rằng tin tức giả thậm chí có thể nhận được sự chú ý nhiều hơn trên mạng xã hội so với tin tức thật.

Chống lại tin giả đòi hỏi những chiến lược chặt chẽ. Tự điều chỉnh và quy định hiệu quả của các nền tảng truyền thông xã hội và công cụ tìm kiếm là quan trọng. Không gian thông tin cần được đổ đầy tin tức chính xác để thay thế tin tức giả, và mỗi cá nhân đều cần trách nhiệm trong việc đối mặt và kiểm soát thông tin mà họ chia sẻ.

Các chiến lược chống lại tin giả đang được nghiên cứu tích cực, từ thuật toán sàng lọc tin tức đến sự tăng cường khả năng phát hiện thông tin sai lệch. Tuy nhiên, để đối mặt với quy mô lớn và sự đa dạng của tin giả, cần có sự kết hợp của lý trí, phương pháp khoa học, và kỹ năng tư duy phản biện.

Nội dung dự án sẽ sử dụng tập dữ liệu news.csv, bao gồm tiêu đề, văn bản và nhãn cho biết tin tức là thật hay giả. Qua đó, chúng ta hy vọng có thể xây dựng một hệ thống hiệu quả để nhận diện và ngăn chặn sự lan truyền của tin tức giả trên mạng.


Giới thiệu về project :
Project với mục đích phát hiện các tin tức giả mạo, bằng bằng cách sừ dụng sklearn chúng ta sẽ xây dựng một  TfidfVectorizer  trên tập dữ liệu của mình. Sau đó, khởi tạo một bộ phân loại thụ động (PassiveAggressive) sao cho phù hợp với mô hình training. Cuối cùng, để kiểm tra tính hiệu quả của mô hình ta sẽ sử dụng accuracy score (chỉ số đánh giá độ chính xác của mô hình máy học)và confusion matrix (là một công cụ quan trọng để đánh giá hiệu suất của mô hình phân loại. Nó thường được sử dụng để hiển thị số lượng các dự đoán đúng và sai trong mỗi phân khúc của tập dữ liệu)để đánh giá.
Để tiếp tục project này, chúng ta cần tìm hiểu về các thuật ngữ Tfidfvectorizer và PassiveAggressive Classifier


1. TfidfVectorizer:

Mô Tả:

TfidfVectorizer giúp chuyển đổi văn bản thành vectơ đặc trưng dựa trên phương pháp TF-IDF. Trong quá trình này, nó tính toán tầm quan trọng của từng từ trong mỗi văn bản và tập trung vào những từ quan trọng đặc biệt đó.

Tầm Quan Trọng:

Term Frequency (TF): Đo lường tần suất xuất hiện của một từ trong một văn bản cụ thể.
Inverse Document Frequency (IDF): Đo lường tầm quan trọng của một từ trên toàn bộ tập dữ liệu, giúp giảm trọng số cho các từ phổ biến như "and," "the,"...

Sử Dụng Trong Dự Án:

Trong dự án phát hiện tin tức giả mạo, TfidfVectorizer giúp biến đổi mỗi tin tức thành một vectơ đặc trưng. Các giá trị trong vectơ này sẽ phản ánh tầm quan trọng của từng từ trong tin tức đối với toàn bộ tập dữ liệu.

2. PassiveAggressiveClassifier:

Mô Tả:

PassiveAggressiveClassifier là một thuật toán máy học linh hoạt và phù hợp cho việc phân loại dữ liệu dòng đến. Điều đặc biệt là nó có khả năng thích ứng với dữ liệu mới mà không cần retraining đầy đủ.

Tính Chủ Động và Bất Chủ Động:

Thuật ngữ "Passive" và "Aggressive" đề cập đến cách mà mô hình xử lý dữ liệu mới. Nó có khả năng "bất chủ động" khi đúng (không cần thay đổi) và "chủ động" khi sai (cập nhật mô hình để phản ánh dữ liệu mới).

Sử Dụng Trong Dự Án:

Trong dự án của bạn, PassiveAggressiveClassifier sẽ là bộ phân loại chính dựa trên vectơ đặc trưng được tạo ra bởi TfidfVectorizer. Nó không chỉ có khả năng phân loại mà còn có thể thích ứng với sự thay đổi trong dữ liệu theo thời gian, điều này là quan trọng trong môi trường phát hiện tin tức giả mạo.
