### 5. Khởi tạo TfidfVectorizer và thực hiện fit, transform
Trong quá trình xử lý văn bản, chúng tôi sử dụng <code>TfidfVectorizer</code> để chuyển đổi các văn bản thành ma trận <code>TF-IDF</code>. Dưới đây là quá trình thực hiện:


```python
# Khởi tạo TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words="english", max_df=0.7)

# Fit and transform train set, transform test set
tfidf_train = tfidf_vectorizer.fit_transform(x_train)
tfidf_test = tfidf_vectorizer.transform(x_test)
```

•	Đầu tiên, khởi tạo một đối tượng <code>TfidfVectorizer</code> với một số tham số, bao gồm <code>stop_words</code> để loại bỏ từ ngữ không quan trọng (ví dụ: "is", "and", "the") và <code>max_df</code> để loại bỏ các từ có tần suất xuất hiện quá cao.

•	Sau đó, sử dụng phương thức <code>fit_transform</code> để huấn luyện và chuyển đổi tập dữ liệu huấn luyện (<code>x_train</code>). Chúng tôi cũng sử dụng <code>transform</code>, để áp dụng cùng bộ biến đổi cho tập dữ liệu kiểm tra (<code>x_test</code>).


### 6. Khởi tạo PassiveAggressionClassifier và thực hiện dự đoán
Sau khi đã chuyển đổi văn bản thành ma trận <code>TF-IDF</code>, chúng tôi sử dụng mô hình <code>PassiveAggressiveClassifier</code> để phân loại tin tức giả và thật. Dưới đây là quá trình thực hiện:

```python
# Khởi tạo PassiveAggressiveClassifier
pac = PassiveAggressiveClassifier(max_iter=50)
pac.fit(tfidf_train, y_train)

# Dự đoán trên tập test và tính toán độ chính xác
y_pred = pac.predict(tfidf_test)
score = accuracy_score(y_test, y_pred)
print(f"Accuracy: {round(score*100,2)}%")
```

•	Chúng tôi khởi tạo một mô hình <code>PassiveAggressiveClassifier</code>, với <code>max_iter</code> là số lượng lần lặp cho quá trình đào tạo.

•	Sau đó, sử dụng phương thức <code>fit</code> để đào tạo mô hình trên tập dữ liệu huấn luyện (<code>tfidf_train</code>) và nhãn tương ứng (<code>y_train</code>)

•	Tiếp theo, chúng tôi thực hiện dự đoán trên tập dữ liệu kiểm tra (<code>tfidf_test</code>) và sử dụng <code>accuracy_score</code> để đánh giá độ chính xác của mô hình.

•	Kết quả sẽ được in ra màn hình, ví dụ: "Accuracy: 95.98%". Điều này cho biết mô hình đạt được độ chính xác khoảng 95.98% trên tập dữ liệu kiểm tra.
