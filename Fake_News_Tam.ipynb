#### 2. Đọc dữ liệu vào DataFrame

Ở bước kế tiếp sau khi đã có được các thư viện cần thiết. Chúng ta sẽ bắt đầu làm việc với nó, đầu tiên là thực hiện các lệnh đọc và in ra dữ liệu theo cách dưới đây:

```python
# Đọc dữ liệu vào dataframe
df = pd.read_csv("./news.csv")
# Xem kích thước và 5 dòng đầu của dữ liệu
print("Shape:", df.shape)
df.head(5)
```

Dữ liệu của chúng ta sẽ bao gồm id, tiêu đề(title), tác giả(author), nội dung(text), và nhãn đánh dấu fake hoặc real (label).
Trong đó nhãn đánh đấu của chúng ta sẽ bao gồm 0 và 1 với nhãn 0 là tin chuẩn (real) và nhãn 1 là tin chưa chuẩn (fake).

#### 3. Thay đổi và lấy các nhãn của DataFrame

Kế đến, sau khi đã làm việc với dữ liệu một cách chuẩn chỉ với id, title, author và label thì chúng ta sẽ thực hiện bước cho label hiển thị trực tiếp "REAL" hoặc "FAKE" thay vì 0 hoặc 1. Vì nếu như chúng ta để nguyên label là 0 hoặc 1 thì có thể người sử dụng sẽ không phân biệt được đâu là REAL hoặc FAKE vì các họ không biết 0 và 1 cái nào là REAL cái nào là FAKE. Nếu giữ 0 và 1 làm output thì phải chú thích cho người dùng hiểu được thông qua một cách nhắc nhở nào đó, điều này rất bất tiện. Do đó, việc thay đổi label là cần thiết.

```python
# Thay đổi các nhãn của dataframe
df["label"] = df["label"].astype("object")
df.loc[(df["label"] == 1), ["label"]] = "FAKE"
df.loc[(df["label"] == 0), ["label"]] = "REAL"
```

```python
# Xem lại 5 dòng đầu của dữ liệu
df.head(5)
```

Ta có thể thấy, label đã được thay đổi output thành REAL hoặc FAKE thuận tiện hơn trong việc sử dụng

Nếu như chỉ muốn xem tin chuẩn hay không chuẩn mà không quan tâm đến nội dung hay tác giả, ta có thể thực hiện như sau:

```python
# Lấy các nhãn từ dataframe
labels = df.label
labels.head(5)
```

#### 4. Chia bộ dữ liệu thành các tập train và test

```python
# Trộn tiêu đề, tác giả và nội dung
df["combined_info"] = df.apply(
    lambda row: f"Title: {row['title']}\nAuthors: {row['author']}\nText:\n{row['text']}",
    axis=1,
)

# Chia bộ dữ liệu
x_train, x_test, y_train, y_test = train_test_split(
    df["combined_info"], labels, test_size=0.2, random_state=7
)
```
